**Skewness and Kurtosis**

Skewness:- 
While we first notice our distribution we check whether our distribution has one mode(peak) or not. If the data is unimodal(has one peak), we check whether it symmetrical or not. If the bulk of our data is at left and the tail is at right we say that our distribution or variable or data is Skewed right or Positively Skewed. If the bulk is at right and the tail is at left we say that our distribution or variable or data is skewed left or negatively skewed.

Kurtosis:- 
Like skewness, kurtosis is a statistical measure that is used to describe the distribution. Whereas skewness differentiates extreme values in one versus the other tail, kurtosis measures extreme values in either tail.
Kurtosis is sometimes confused with a measure of the peakedness of a distribution. However, kurtosis is a measure that describes the shape of a distribution's tails in relation to its overall shape.

#Multivariate Analysis

While making a machine learning model, there are 2 basic things that we need to know.

1. Nature of data or an individual variable.
2. Impact of an individual variable over other variables. 

While univariate analysis involves deals with finding the nature of a  single variable, Multivariate analysis, deals with investigating the impact of individual variable on other variables. 

Best way to find what impact does an individual variable does over other variables is to visualize our dataset. 

**Correlation Matrix Plot**

The best way to find the impact of one variable over the other variable is to find the correlation. 

Correlation gives an indication on how the changes in one variable are related to the other variable. If two variables have a positive correlation they will change in the same direction. If two variables have a negative correlation they will change in different direction. if the correlation is zero, then both the variables are not related to each other.

An easier way to understand this is through direct and inverse variation.
1. Two quantities are said to be in direct variation if the increase in one also increases the other. Hence two variables having a positive correlation will have a relationship of direct variation.
2. Two quantities are said to be in inverse variation, if the increase in one causes decrease the other. Hence two variables having a negative correlation will have a relationship of inverse variation.
3. If the two variables are have neither the direct variation nor inverse variation then the two variables are not related to each other and hence have a correlation of zero.


We will now go ahead and try to understand this concept a correlation Matrix plot.

```
# Plotting the correlation matrix
Data.corr()
```

```
# visualizing the correlation matrix in a graphical form.
correlations= Data.corr()

ax = sns.heatmap(correlations,vmin=-1, vmax=1, center=0,cmap=sns.diverging_palette(20, 220, n=200),square=True)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45,horizontalalignment='right');
```

Now once we have plotted the Correlation Plot, we will now try to understand what this correlation plot speaks about our data. 

The most notable thng is the diagonal which is same colour. If we take a look at color bar at the right we see that the Dark blue has a number 1.0 , and the color dark red/brown as -1.0, Which meanns the variables with darker shades of blue has positive correlation, and the variables with the darker shades of red/brown are negatively correlated. Variables having shades of white are not related to each other
